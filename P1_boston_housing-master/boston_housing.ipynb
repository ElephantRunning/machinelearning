### 编程练习3：定义衡量标准
如果不能对模型的训练和测试的表现进行量化地评估，我们就很难衡量模型的好坏。通常我们会定义一些衡量标准，这些标准可以通过对某些误差或者拟合程度的计算来得到。在这个项目中，你将通过运算[*决定系数*](http://stattrek.com/statistics/dictionary.aspx?definition=coefficient_of_determination) R<sup>2</sup> 来量化模型的表现。模型的决定系数是回归分析中十分常用的统计信息，经常被当作衡量模型预测能力好坏的标准。

R<sup>2</sup>的数值范围从0至1，表示**目标变量**的预测值和实际值之间的相关程度平方的百分比。一个模型的R<sup>2</sup> 值为0还不如直接用**平均值**来预测效果好；而一个R<sup>2</sup> 值为1的模型则可以对目标变量进行完美的预测。从0至1之间的数值，则表示该模型中目标变量中有百分之多少能够用**特征**来解释。_模型也可能出现负值的R<sup>2</sup>，这种情况下模型所做预测有时会比直接计算目标变量的平均值差很多。_

- 将他们的表现评分储存到`score`变量中。

手工计算判定系数(R Square)时，判定系数的计算公式为：

```python
# TODO 3 可选

# 不允许导入任何计算决定系数的库

def performance_metric2(y_true, y_predict):
    """手工计算判定系数(R Square)"""
    
    # 实际值的均值
    y_bar = np.mean(y_true)
    y_bar_vector = pd.Series(np.full(len(y_true), y_bar))

    # 计算SSE
    true_minus_pre = y_true - y_predict
    true_minus_pre = true_minus_pre ** 2
    SSE = np.sum(true_minus_pre)

    # 计算SST
    true_minus_ybar = y_true - y_bar_vector
    true_minus_ybar = true_minus_ybar ** 2
    SST = np.sum(true_minus_ybar)

    # 计算判定系数
    score = 1 - SSE/SST
    return score

print "R Square2:", performance_metric2(y_test, reg.predict(X_test))
```

## 第五步. 选择最优参数

### 问题 7- 网格搜索（Grid Search）
*什么是网格搜索法？如何用它来优化模型？*

### 问题 7 - 回答:
网格搜索根据给定的模型自动进行交叉验证，不断尝试可能的交叉验证子集数量(k值)并依此计算模型的预测(泛化)能力，多次尝试后最终确定最优的k参数，使模型达到最优的预测能力。<br>其网格的含义在于：在使用网格搜索时，以字典形式指定待搜索优化参数列表，对于每一个参数，均使用K折交叉验证计算评分函数值并最终计算均值，待每个参数均计算完毕后，依据交叉验证评分(最大值或最小值)给出最优参数；在搜索最优参数过程中，参数列表及K折交叉验证如同数据表格的行名称、列名称，形似网格，逐个按照网格计算评分值及依据计算结果搜索给出最优参数，故称之为网格搜索。<br>
网格搜索的是在我们给定参数列表(“param_grid”形参)的情况下，结合K折交叉验证计算一个算法所有可能的参数组合，依据评分结果反馈出最优参数。<br>
网格搜索评分函数(“scoring”形参)默认为None，一般需我们指定评分函数，以评分函数作为K折交叉验证评分的工具和依据；如果不指定评分函数，GridSearchCV将使用模型默认的评分方法。

- [GridSearchCV](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html)中的`'cv_results_'`属性能告诉我们什么？

### 问题 8 - 回答：
1. K折交叉验证，将训练集分割成K个子样本，一个单独的子样本被保留作为验证模型的数据，其他K-1个样本用来训练。交叉验证重复K次，每个子样本验证一次，平均K次的结果或者使用其它结合方式，最终得到一个单一估测。默认情况下，shuffle=False，random_state形参不生效，KFold 对数据顺序切分(顺序切成K份)；当shuffle == True时，random_state形参生效，此时根据random_state实参值对数据随机切分。
2. GridSearchCV实现了fit，predict，predict_proba等方法，并通过交叉验证对参数空间进行求解，寻找最优化的结果和参数。
3. 'cv_results\_'属性包含二维数据表格(包含列名称及列内容)形式的不同参数的组合方式及其计算结果，以便于分析K折交叉验证的结果，从而确定预测模型达到最优时的参数；该结果可以导入到DataFrame中或者存储为'csv'格式的电子表格。    
4. 网格搜索时如果不使用交叉验证意味着只对样本做1次划分，设定比例将其分为唯一的训练集和测试集，在训练集上建立模型，进而利用该模型在测试集上进行性能评估，由于训练得到的模型及验证次数太少(均为1)，评估结果存在偶然性，可信程度不高，进而影响模型最终确定最优参数；交叉验证将样本随机分为K份，建模算法运行K次并进行K次评估后才能得出K折的性能指标，并且对每一个参数均重复同样的计算，极大降低了计算结果的偶然性、提高了其可信程度，使网格搜索得出的最优参数可信。

### 编程练习 4：训练最优模型

在下方 `fit_model` 函数中，你需要做的是：

3. **定义 `'params'` 变量**: 为 `'max_depth'` 参数创造一个字典，它的值是从1至10的数组;
4. **定义 `'scoring_fnc'` 变量**: 使用 `sklearn.metrics` 中的 [`make_scorer`](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.make_scorer.html)  创建一个评分函数；
 将 `‘performance_metric’` 作为参数传至这个函数中；
5. **定义 `'grid'` 变量**: 使用 `sklearn.model_selection` 中的 [`GridSearchCV`](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html) 创建一个网格搜索对象；将变量`'regressor'`, `'params'`, `'scoring_fnc'`和 `'cross_validator'` 作为参数传至这个对象构造函数中；

```python
# TODO 6
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, KFold, GridSearchCV

data = pd.read_csv('bj_housing.csv')
prices = data['Value']
features = data.drop('Value', axis=1)

# 完成
print "Beijing housing dataset has {} data points with {} variables each.".format(*data.shape)

# 数据封装成数组
features = np.array(features)
prices = np.array(prices)
# 目标：计算价值的最小值
minimum_price = np.min(prices)
# 目标：计算价值的最大值
maximum_price = np.max(prices)
# 目标：计算价值的平均值
mean_price = np.mean(prices)
# 目标：计算价值的中值
median_price = np.median(prices)
# 目标：计算价值的标准差
std_price = np.std(prices)

# 目标：输出计算的结果
ts = 10000
print "Statistics for Beijing housing dataset:\n"
print "Minimum price: ￥{:,.2f}".format(minimum_price * ts)
print "Maximum price: ￥{:,.2f}".format(maximum_price * ts)
print "Mean price: ￥{:,.2f}".format(mean_price * ts)
print "Median price ￥{:,.2f}".format(median_price * ts)
print "Standard deviation of prices: ￥{:,.2f}".format(std_price * ts)

# todo2 数据分割与重排
X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size=0.2, random_state=1,shuffle=True)


def performance_metric(XX_test, yy_test):
    """计算并返回预测值相比于预测值的分数"""
    score = reg.score(XX_test, yy_test)
    return score


def performance_metric2(y_true, y_predict):
    """计算并返回预测值相比于预测值的分数
       使用1-SSE/SST计算，与SKLearn中的算法一致
    """

    # 实际值的均值
    y_bar = np.mean(y_true)
    # 预测值的均值
    # y_bar = np.mean(y_predict)
    y_bar_vector = pd.Series(np.full(len(y_true), y_bar))

    # 计算SSE
    true_minus_pre = y_true - y_predict
    true_minus_pre = true_minus_pre ** 2
    SSE = np.sum(true_minus_pre)

    # 计算SST
    true_minus_ybar = y_true - y_bar_vector
    true_minus_ybar = true_minus_ybar ** 2
    SST = np.sum(true_minus_ybar)

    # 计算判定系数
    score = 1 - SSE / SST

    return score


# 训练得到回归模型，进行预测
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)

print "R square:", performance_metric(X_test, y_test)
# 手工计算判定系数
print "R square2:", performance_metric2(y_test, reg.predict(X_test))


# 网格搜索
def fit_model(X, y):
    """ 基于输入数据 [X,y]，利于网格搜索找到最优的决策树模型"""
    from sklearn.tree import DecisionTreeRegressor
    from sklearn.metrics import make_scorer

    cross_validator = KFold(n_splits=10, shuffle=True, random_state=1)
    regressor = DecisionTreeRegressor(random_state=1)
    params = {"max_depth": range(1, 11)}
    scoring_fnc = make_scorer(performance_metric2)
    grid = GridSearchCV(estimator=regressor, param_grid=params, scoring=scoring_fnc,
                        cv=cross_validator)

    # 基于输入数据 [X,y]，进行网格搜索
    grid = grid.fit(X, y)

    # 返回网格搜索后的最优模型
    return grid.best_estimator_

optimal_reg = fit_model(X_train, y_train)
print "Optimal model info:", optimal_reg
print "The score of optimal model:", optimal_reg.score(X_test,y_test)
```
