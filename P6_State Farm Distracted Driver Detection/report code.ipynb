```python
def show_image_count_each_class():
    """
    显示各分类图片数量

    Returns:
        无
    """
    csv_data = dpu.read_csv_file()
    grouped_imgs = csv_data.groupby('classname')
    data_list = []
    for class_name in dpu.TOTAL_LABELS:
        data_list.append(len(grouped_imgs.get_group(class_name)))
    data_raw = {
        "classes":dpu.TOTAL_LABELS,
        "count":data_list
    }
    data = pd.DataFrame(data_raw)
    plt.figure()
    x=np.arange(10)+1
    y=np.array(list(data['count']))
    xticks1=list(data["classes"]) 
    plt.bar(x,y,width = 0.7,align='center',color = 'b',alpha=0.8)
    plt.xticks(x, xticks1)
    plt.xlabel('classes')
    plt.ylabel('image count')
    for a,b in zip(x,y):
        plt.text(a, b+0.05, '%.0f' % b, ha='center', va='bottom', fontsize=7)
    plt.ylim(0,2600)
    plt.show()
show_image_count_each_class()
```

```python
def img_count_by_driver_ID():
    """
    统计各驾驶者ID的图像数量

    Returns:
        无
    """
    csv_data = dpu.read_csv_file()
    drive_IDs = dpu.sorted_unique_list(csv_data["subject"])
    drive_ID_count = []
    for id in drive_IDs:
        drive_ID_count.append(len(csv_data[csv_data.subject == id].index.tolist()))
    plt.figure(figsize=(17,10))
    vip.draw_bar(plt, drive_IDs, drive_ID_count, show_text=True)
    plt.xlim(0,27)
    plt.show()

img_count_by_driver_ID()
```

```python
def read_csv_split_driver_ID():
    """
    按照驾驶者ID统计不同的分类数据

    Returns:
        无
    """
    csv_data = dpu.read_csv_file()
    drive_IDs = dpu.sorted_unique_list(csv_data["subject"])
    sub_plot_x_count = 5
    sub_plot_y_count = 6
    fig, sub_plt = plt.subplots(sub_plot_x_count, sub_plot_y_count, figsize=(17,10))
    index = 0
    for drive_ID in drive_IDs:
        p_class_index = csv_data[csv_data.subject == drive_ID].index.tolist()
        sub_data = csv_data[p_class_index[0]:(p_class_index[len(p_class_index) - 1] + 1)]
        class_names = dpu.sorted_unique_list(sub_data["classname"])
        class_count = []
        for c_n in class_names:
            class_count.append(len(sub_data[sub_data.classname == c_n].index.tolist()))
        d1 = int(index / sub_plot_y_count)
        d2 = int(index % sub_plot_y_count)
        vip.draw_bar_in_subplot(sub_plt[d1][d2], class_names, class_count, "The class' count of {}".format(drive_ID), True)
        index += 1
    for i in range(sub_plot_x_count*sub_plot_y_count - index):
        d1 = int(index / sub_plot_y_count)
        d2 = int(index % sub_plot_y_count)
        vip.hide_subplot(sub_plt[d1][d2])
        index += 1
    plt.tight_layout(h_pad=0.1)
    plt.subplots_adjust(wspace=0.3, hspace=0.5)
    plt.show()
read_csv_split_driver_ID()
```

## 通用迁移学习模型——K折交叉法、多GPU

```python
class ParallelModelCheckpoint(Callback):
    """
    每个Epoch结束后保存训练得到的模型
    """
    def __init__(self, model_train, model_save_prefix, num_folds):
        self.model_to_save = model_train
        self.model_prefix = model_save_prefix
        self.num_folds = num_folds
    def on_epoch_end(self, epoch, logs=None):
        self.model_to_save.save('{}{}_{}.h5'.format(self.model_prefix, self.num_folds, epoch))
```

```python
def transferl_train_save_model_kfold_multi_GPU(tf_model_class, optimizer, lower_unlock, higher_unlock, n_splits, batch_size, train_epochs, img_shape, train_size=0.8, gpus=4, splice_train_images=True):
    """
    使用K折交叉法划分驾驶者ID，在多块GPU上训练迁移学习分类模型并保存训练结果

    Args:
        tf_model_class: function, 迁移学习函数名称
        optimizer: function or str, 优化器函数或名称
        lower_unlock: int, 该值以下的迁移学习模型层将被解锁并训练
        higher_unlock: int, 该值以上的迁移学习模型层将被解锁并训练
        n_splits: int, 交叉验证的次数
        batch_size: int, 训练时每次读取的样本数量
        train_epochs: int, 每折训练轮数
        img_shape: (int,int,int), 图像输入尺寸元组
        train_size: float, (0,1)之间的浮点数，训练集所占比例
        gpus: int, 用于训练的GPU数量，默认为4
        splice_train_images: bool, 是否左右拼接训练集图片，将训练集扩展为原来的2倍，默认为True
    Returns:
        str: 保存结果的目录名称
    """
    csv_data = dpu.read_csv_file()
    csv_data = csv_data.sample(frac=1, random_state=4)
    driver_ids = csv_data['subject']
    x_train_names_all = np.array(csv_data[dpu.CLASSNAME_IMGNAME])
    y_train_labels_all = np.array(csv_data['classname'])
    _, driver_indices = np.unique(np.array(driver_ids), return_inverse=True)
    
    modle_name = gu.get_name_of_obj(globals(), tf_model_class)
    gu.p('modle_name', modle_name)
    res_path, res_name = dpu.mkdir_results(modle_name)
    gu.p('res_dir_name', res_name)
    rs = GroupShuffleSplit(n_splits=n_splits, test_size=None, train_size=train_size, random_state=4)
    num_folds = 0
    for train_indices, val_indices in rs.split(x_train_names_all, y_train_labels_all, groups=driver_indices):
        print('Fold {}/{}......'.format(num_folds + 1, n_splits))
        x_train_names, y_train = x_train_names_all[train_indices,...], y_train_labels_all[train_indices,...]
        x_val_names, y_val = x_train_names_all[val_indices,...], y_train_labels_all[val_indices,...]
        if splice_train_images:
            X_train, y_train = dpu.splice_img_left_right(x_train_names, y_train, img_shape)
        else:
            X_train, y_train = dpu.generate_features_labels_array(x_train_names, y_train, img_shape)
        X_val, y_val = dpu.generate_features_labels_array(x_val_names, y_val, img_shape)
        
        gu.p('X_train.shape', X_train.shape)
        gu.p('y_train.shape', y_train.shape)
        gu.p('x_val_names.shape', x_val_names.shape)
        gu.p('y_val.shape', y_val.shape)

        start_time_ini = time.time()
        with tf.device('/cpu:0'):
            x = Input(shape=img_shape)
            base_model = tf_model_class(include_top=False, input_tensor=x, input_shape=img_shape,weights='imagenet', pooling='avg')
            for layer in base_model.layers:
                layer.trainable = False
            y = Dropout(0.2)(base_model.output)
            y = Dense(10, activation='softmax', kernel_initializer='he_normal')(y)
            model = Model(inputs=base_model.input, outputs=y, name='Transfer_Learning')
        for layer in model.layers[:lower_unlock]:
            layer.trainable = True
        for layer in model.layers[lower_unlock:higher_unlock]:
            layer.trainable = False
        for layer in model.layers[higher_unlock:]:
            layer.trainable = True
        parallel_model = multi_gpu_model(model, gpus=gpus)
        parallel_model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
        model.save(os.path.join(res_path, '{}lock_{}_{}_{}.h5'.format(dpu.M_COMPILE_PREFIX, lower_unlock, higher_unlock, num_folds)))

        # 动态调整学习率
        def learning_rate(epoch):
            ini_lr = 0.0002
            lr = ini_lr * pow(10, -epoch)
            return lr
        
        log_path = os.path.join(res_path, '{}_{}'.format(dpu.TENSORBOARD, num_folds))
        fou.mkdir(log_path)
        callbacks = [
            EarlyStopping(monitor='val_loss', patience=5, verbose=0, mode='min'),
            ParallelModelCheckpoint(model, os.path.join(res_path, dpu.M_CHECKPOINT_PREFIX), num_folds),
            TensorBoard(log_dir=log_path, histogram_freq=0),
            LearningRateScheduler(learning_rate)
        ]
        parallel_model.fit(X_train,
                           y_train,
                           batch_size=batch_size,
                           epochs=train_epochs,
                           shuffle=True,
                           verbose=1,
                           validation_data=(X_val, y_val),
                           callbacks=callbacks)
        print('Flod {} trainning during time(s):{:.2f}'.format(num_folds + 1, gu.get_during_time(start_time_ini)))
        num_folds += 1
    return res_name
```

## 结果可视化

```python
def grad_cam(input_model, image, cls, layer_name, target_size):
    """
    使用grad_cam方法可视化图片

    Args:
        input_model: model, 模型
        image: numpy.array, 图片数组
        cls: str, 图片分类
        target_size: (int,int), 缩放后的图片大小
    Returns:
        numpy.array: 图片数组
    """
    y_c = input_model.output[0, cls]
    conv_output = input_model.get_layer(layer_name).output
    grads = K.gradients(y_c, conv_output)[0]
    gradient_function = K.function([input_model.input], [conv_output, grads])
    output, grads_val = gradient_function([image])
    output, grads_val = output[0, :], grads_val[0, :, :, :]
    weights = np.mean(grads_val, axis=(0, 1))
    cam = np.dot(output, weights)
    cam = cv2.resize(cam, target_size, cv2.INTER_LINEAR)
    cam = np.maximum(cam, 0)
    cam = cam / cam.max()
    return cam

def compute_and_show_grad_cam(model, img_path, target_size, layer_name='block5_conv3'):
    """
    使用grad_cam方法对图片进行可视化处理并显示结果

    Args:
        input_model: model, 模型
        img_path: str, 图片路径名称
        target_size: (int,int), 缩放后的图片大小
        layer_name: 模型最后一个卷积层名称
    Returns:
        无
    """

    img_raw = vip.read_image_file(img_path, target_size=target_size)
    img_preprocessed = np.expand_dims(img_raw, axis=0)
    img_preprocessed = xception.preprocess_input(img_preprocessed)

    predictions = model.predict(img_preprocessed)
    p_list = predictions[0].tolist()
    cls = p_list.index(max(p_list))
    gu.p('Predict class name', cls)
    gradcam = grad_cam(model, img_preprocessed, cls, layer_name, target_size)

    plt.axis('off')
    plt.imshow(vip.read_image_file(img_path, target_size=target_size,dtype_float32=False))
    plt.imshow(gradcam, cmap='jet', alpha=0.5)
    plt.show()
```
